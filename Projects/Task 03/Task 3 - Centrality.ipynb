```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

# 1 - Mean eigenvector centrality of `USairport500` network

```python
G = nx.read_edgelist("/content/drive/MyDrive/Colab Notebooks/Complex Networks/Project 3/data/USairport500.txt", nodetype=int)
```

```python
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
EC = dict(nx.eigenvector_centrality(G, max_iter = 1000))
#print('Eigenvetor centrality:', EC)
print('Mean Eigenvetor centrality:', np.mean(list(EC.values())))
```

# 2 - Pearson correlation for betweeeness centrality and degree of `hamsterster` network

```python
G = nx.read_edgelist("/content/drive/MyDrive/Colab Notebooks/Complex Networks/Project 3/data/hamsterster.txt", nodetype=int)
```

```python
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
B = list(nx.betweenness_centrality(G).values())
B = np.array(B)

vk = dict(G.degree())
vk = list(vk.values())

Bk = list()
ks = list()
for k in np.arange(np.min(vk), np.max(vk)+1):
    aux = vk == k
    if(len(B[aux]) > 0):
        av_B = np.mean(B[aux]) #average clustering among all the nodes with degree k
        Bk.append(av_B)
        ks.append(k)

from scipy.stats import pearsonr
# calculate Pearson's correlation
corr, _ = pearsonr(ks, Bk)
print('Pearsons correlation: %.3f' % corr)
```

# 3 - Spearman coefficient for closeness centrality and k-core for `Jazz` network

```python
G = nx.read_edgelist("/content/drive/MyDrive/Colab Notebooks/Complex Networks/Project 3/data/jazz.txt", nodetype=int)
```

```python
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
KC= dict(nx.core_number(G))
#print('k-core:', KC)
KC = list(KC.values())

CLC = dict(nx.closeness_centrality(G))
#print('Closeness centrality', CLC)
CLC = list(CLC.values())

from scipy.stats import spearmanr
# calculate Spearman 's correlation
corr, _ = spearmanr(KC, CLC)
print('Spearman  correlation: %.3f' % corr)
```
